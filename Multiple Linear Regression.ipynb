```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
std_reg = LinearRegression()
std_reg.fit(x_train, y_train)
```

```python
y_predict = std_reg.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
sns.regplot(x=y_test, y=y_predict, ci=4, color="blue")
plt.plot(x_train,std_reg.predict(x_train),color="pink")
plt.title("Breast Cancer Analysis")
plt.xlabel("radius_mean" "texture_mean" "perimeter_mean" "area_mean" "smoothness_mean" "compactness_mean" "concavity_mean" "concave points_mean" "symmetry_mean" "fractal_dimension_mean" "radius_se" "texture_se" "perimeter_se" "area_se" "smoothness_se" "compactness_se" "concavity_se" "concave points_se" "symmetry_se" "fractal_dimension_se" "radius_worst" "texture_worst" "perimeter_worst" "area_worst" "smoothness_worst" "compactness_worst" "concavity_worst" "concave points_worst" "symmetry_worst")
plt.ylabel("fractal_dimension_worst")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, y_predict)
print(MAE)
```

```python
MSE = mean_squared_error(y_test, y_predict)
print(MSE)
```

```python
RMSE = np.sqrt(MSE)
print(RMSE)
```

```python
LRMSE = np.log(RMSE)
print(LRMSE)
```
